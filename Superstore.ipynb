# Distribution of Numerical Values

## Setting graph dimensions:

```python
sns.set(rc={"figure.figsize":(8, 4)})
```

## Sales:

```python
sns.distplot(df['Sales'], kde=False, color='blue', bins=50)
plt.title('Distrubution of Sales', fontsize=18)
plt.xlabel('Sales', fontsize=16)
plt.ylabel('Density', fontsize=16)
```

## Profit:

```python
sns.distplot(df['Profit'], kde=False, color='blue', bins=50)
plt.title('Distrubution of Profit', fontsize=18)
plt.xlabel('Sales', fontsize=16)
plt.ylabel('Density', fontsize=16)
```

## Quantity:

```python
sns.distplot(df['Quantity'], kde=False, color='blue', bins=50)
plt.title('Distrubution of Quantity', fontsize=18)
plt.xlabel('Sales', fontsize=16)
plt.ylabel('Density', fontsize=16)
```

## Discount:

```python
sns.distplot(df['Discount'], kde=False, color='blue', bins=50)
plt.title('Distrubution of Discount', fontsize=18)
plt.xlabel('Sales', fontsize=16)
plt.ylabel('Density', fontsize=16)
```

# Distribution of Categorical columns

## Ship mode:

```python
sns.countplot(data = df,x = 'Ship Mode')
plt.title('Distrubution by Ship mode', fontsize=18)
plt.xlabel('Ship mode', fontsize=16)
plt.ylabel('Count', fontsize=16)
```

## Segment:

```python
sns.countplot(data = df,x = 'Segment')
plt.title('Distrubution by Segment', fontsize=18)
plt.xlabel('Segment', fontsize=16)
plt.ylabel('Count', fontsize=16)
```

## Region:

```python
sns.countplot(data = df,x = 'Region')
plt.title('Distrubution by Region', fontsize=18)
plt.xlabel('Region', fontsize=16)
plt.ylabel('Count', fontsize=16)
```

## Category:

```python
sns.countplot(data = df,x = 'Category')
plt.title('Distrubution by Category', fontsize=18)
plt.xlabel('Category', fontsize=16)
plt.ylabel('Count', fontsize=16)
```

# Sales with respect to Geographical features

## Total Sales:

### By region

```python
sns.barplot(x=df['Region'],y=sum(df.['Sales'], ci=None)
plt.title('Total Sales by Region', fontsize=18)
plt.xlabel('Region', fontsize=16)
plt.ylabel('Sales', fontsize=16)
```
